#1. Introdução

A experimentação é uma ferramenta fundamental para a produção de
conhecimento científico. Por meio de experimentos planejados, é possível
investigar fenômenos, testar hipóteses e avaliar o efeito de diferentes
condições sobre uma determinada resposta. Quanto melhor for o
planejamento experimental, maior será a qualidade dos resultados obtidos
e das conclusões derivadas do estudo.

Neste notebook serão apresentados os principais conceitos relacionados
ao planejamento experimental, incluindo tratamentos, fatores, níveis,
unidades experimentais, variáveis resposta, erro experimental e os
princípios básicos da experimentação.

## 1.1. O que é um experimento?

Um experimento é um procedimento planejado no qual determinadas
condições são controladas e observadas para avaliar seus efeitos sobre
uma ou mais variáveis de interesse.

Durante a experimentação, busca-se reduzir a influência de fatores
externos para que as diferenças observadas possam ser atribuídas aos
tratamentos estudados.

# 2. Elementos centrais de um experimento

Todo experimento é composto por alguns elementos fundamentais:

- Tratamento;
- Fator e níveis;
- Unidade experimental;
- Variável resposta;
- Repetições.

A correta definição desses elementos é essencial para garantir a
qualidade dos resultados obtidos.

## 2.1. Tratamento

O tratamento corresponde à condição aplicada ou avaliada durante o
experimento. É o elemento que diferencia os grupos experimentais e que
se deseja comparar ao final da pesquisa.

Os tratamentos podem ser:

### Quantitativos

Possuem valores numéricos ordenáveis.

Exemplos: - Teor de fibras (%); - Temperatura de cura; - Relação
água/aglomerante.

### Qualitativos

Representam categorias sem ordem natural.

Exemplos: - Tipo de fibra; - Tipo de matriz; - Método de tratamento
superficial.

## 2.2. Fator e níveis

O fator é a variável independente que se deseja estudar. Os níveis
correspondem às diferentes condições assumidas por esse fator.

Em outras palavras, o fator representa “o que está sendo investigado”,
enquanto os níveis representam “as condições comparadas”.

### Exemplo

**Fator:** tipo de fibra

**Níveis:** - Juta; - Sisal; - Abacaxi.

## 2.3. Unidade experimental

A unidade experimental é o elemento que recebe diretamente o tratamento
e fornece os dados utilizados na análise estatística.

Ela representa a menor unidade independente do experimento.

### Exemplos

- Um corpo de prova;
- Uma placa cimentícia;
- Um tecido submetido ao ensaio de tração;
- Uma amostra de solo.

## 2.4. Variável resposta

A variável resposta é a característica medida durante o experimento e
que será utilizada para avaliar o efeito dos tratamentos.

É a variável que efetivamente responde às alterações promovidas pelos
fatores estudados.

### Exemplos

- Resistência à tração;
- Resistência à flexão;
- Módulo de elasticidade;
- Absorção de água;
- Tenacidade.

## 2.5. Repetição e erro experimental

Mesmo quando duas unidades experimentais recebem exatamente o mesmo
tratamento, os resultados obtidos dificilmente serão idênticos. Essa
diferença é conhecida como erro experimental.

As repetições são fundamentais porque permitem estimar essa
variabilidade natural e aumentar a confiabilidade das análises
estatísticas.

Quanto maior a variabilidade do material estudado, maior tende a ser a
necessidade de repetições.

# 3. Delineamento experimental

O delineamento experimental corresponde à forma como os tratamentos são
distribuídos entre as unidades experimentais.

Seu principal objetivo é reduzir a influência da variabilidade não
controlada e permitir que as diferenças observadas sejam atribuídas aos
tratamentos estudados.

Além disso, o delineamento define a forma como os dados serão
posteriormente analisados por métodos estatísticos, como a análise de
variância (ANOVA).

## Por que planejar um experimento?

Um experimento bem planejado permite:

- Reduzir erros experimentais;
- Otimizar tempo e recursos;
- Aumentar a confiabilidade dos resultados;
- Obter conclusões estatisticamente válidas;
- Identificar corretamente os efeitos dos tratamentos.

Sem planejamento adequado, mesmo uma grande quantidade de dados pode não
produzir conclusões confiáveis.

## 3.1. Efeito de borda

O efeito de borda ocorre quando a resposta observada em uma unidade
experimental sofre influência das unidades vizinhas ou das condições
existentes nas extremidades da área experimental.

Esse fenômeno é mais comum em experimentos de campo, nos quais fatores
externos, como sombreamento, vento, temperatura ou interferência entre
tratamentos, podem alterar os resultados obtidos.

Para minimizar esse problema, costuma-se utilizar áreas de bordadura ou
faixas de segurança, nas quais os dados não são coletados para análise
estatística.

## 3.2. Os 3 princípios básicos da experimentação

Os delineamentos experimentais são fundamentados em três princípios
básicos:

1.  Casualização (Aleatorização);
2.  Repetição;
3.  Controle Local (Bloqueamento).

Esses princípios permitem reduzir vieses, estimar o erro experimental e
aumentar a eficiência do experimento.

### 3.2.1. Casualização (Aleatorização)

A casualização consiste na distribuição aleatória dos tratamentos entre
as unidades experimentais.

Seu objetivo é evitar que fatores externos influenciem sistematicamente
os resultados do experimento.

Por meio da aleatorização, todos os tratamentos possuem a mesma
probabilidade de serem aplicados em qualquer unidade experimental,
reduzindo possíveis vieses do pesquisador.

### Exemplo

Suponha um estudo com três tipos de tecidos:

- T1 → juta;
- T2 → sisal;
- T3 → abacaxi.

Em vez de ensaiar todos os corpos de prova de um mesmo tratamento em
sequência, realiza-se um sorteio da ordem dos ensaios:

T2 → T1 → T3 → T1 → T2 → T3

Essa estratégia reduz a influência de fatores externos, como variações
de temperatura, umidade ou ajustes do equipamento.

### Experimentos cegos

Em alguns estudos, o conhecimento prévio sobre os tratamentos pode
influenciar, consciente ou inconscientemente, a avaliação dos
resultados. Para reduzir esse tipo de viés, podem ser utilizados
experimentos cegos.

- **Experimento cego:** apenas o pesquisador conhece os tratamentos
  aplicados às unidades experimentais. O avaliador não possui essa
  informação.

- **Experimento duplo-cego:** nem o pesquisador responsável pela
  aplicação dos tratamentos nem o avaliador conhecem a identidade dos
  tratamentos durante a coleta dos dados.

Essas estratégias são amplamente utilizadas em pesquisas médicas,
farmacêuticas e sensoriais, contribuindo para aumentar a imparcialidade
da avaliação.

Em ensaios mecânicos de materiais, como tração, compressão e flexão, os
experimentos cegos são menos frequentes, pois as diferenças entre os
tratamentos geralmente são visíveis ao pesquisador.

Entretanto, em avaliações subjetivas, como inspeções visuais, análise de
fissuração ou ensaios de percepção, a utilização de procedimentos cegos
pode contribuir para reduzir vieses de avaliação.

### 3.2.2. Repetição

A repetição corresponde ao número de unidades experimentais submetidas
ao mesmo tratamento.

Ela é fundamental para estimar a variabilidade natural do experimento e
possibilitar a aplicação de testes estatísticos.

Quanto maior a variabilidade do material, maior tende a ser a
necessidade de repetições.

### Exemplo

Deseja-se avaliar a resistência à tração de tecidos de juta.

Em vez de ensaiar apenas um corpo de prova, realizam-se:

- 10 ensaios para juta;
- 10 ensaios para sisal;
- 10 ensaios para abacaxi.

As repetições permitem estimar a dispersão dos resultados e aumentar a
confiabilidade das conclusões.

### 3.2.3. Controle local (Bloqueamento)

O controle local consiste em agrupar unidades experimentais semelhantes
em blocos, reduzindo a influência de fontes conhecidas de variabilidade.

Essa técnica é utilizada quando existe algum fator externo que pode
afetar os resultados e que não faz parte dos tratamentos estudados.

### Exemplo

Suponha que os ensaios de tração sejam realizados em dois dias
diferentes.

Como pequenas variações ambientais podem ocorrer entre os dias de
ensaio, cada dia pode ser considerado um bloco experimental:

- Bloco 1 → Dia 1;
- Bloco 2 → Dia 2.

Dessa forma, a influência do dia de ensaio é separada do efeito dos
tratamentos.

### *Resumo*

✔ Casualização → reduz vieses.

✔ Repetição → permite estimar o erro experimental.

✔ Controle local → reduz a variabilidade causada por fatores externos.

Juntos, esses três princípios constituem a base do planejamento
experimental moderno.

## 3.3. Etapas do planejamento experimental

### 1 - Definição do problema

Estabelecer claramente a pergunta que o experimento pretende responder.

### 2 - Escolha dos fatores e níveis

Definir quais variáveis serão investigadas e quais condições serão
comparadas.

### 3 - Definição das variáveis resposta

Selecionar as propriedades que serão medidas para avaliar o efeito dos
tratamentos.

### 4 - Escolha do delineamento

Definir a estratégia experimental, número de repetições e forma de
distribuição dos tratamentos.

### 5 - Execução e análise

Realizar os ensaios, organizar os dados e aplicar os métodos
estatísticos adequados.

apresentado um exemplo de planejamento experimental relacionado à
caracterização mecânica de tecidos produzidos com fibras vegetais.

O estudo foi estruturado de forma a identificar os principais elementos

Além disso, o exemplo permite visualizar como os princípios da

## *Caracterização mecânica de tecidos de fibras vegetais*

### Contextualização

As fibras vegetais apresentam elevada variabilidade natural devido a
fatores como origem da planta, idade, condições climáticas e processos
de extração.

Por esse motivo, o planejamento experimental é particularmente
importante em estudos que envolvem a caracterização mecânica desses
materiais.

Neste exemplo, deseja-se comparar o comportamento à tração de tecidos
produzidos com fibras de juta e fibras de abacaxi.

### Problema experimental

Existe diferença significativa entre as propriedades mecânicas dos
tecidos produzidos com fibras de juta e fibras de abacaxi?

### Hipótese

$H_0$: Não existe diferença significativa entre os tecidos avaliados.

$H_1$: Existe diferença significativa entre os tecidos avaliados.

### Definição dos elementos experimentais

### → Tratamentos

Os tratamentos correspondem aos tecidos produzidos com diferentes fibras
vegetais:

- T1 → tecido de juta;
- T2 → tecido de abacaxi.

### → Fator

Tipo de fibra vegetal.

### → Níveis

- Nível 1: juta;
- Nível 2: abacaxi.

### → Unidade experimental

Cada corpo de prova de tecido submetido individualmente ao ensaio de
tração.

### → Variáveis resposta

As propriedades mecânicas obtidas durante o ensaio de tração:

- Resistência máxima à tração (MPa);
- Deformação na ruptura (%);
- Módulo de elasticidade (GPa);
- Tenacidade (área sob a curva tensão × deformação).

### Número de repetições

Como materiais vegetais apresentam elevada variabilidade natural,
recomenda-se utilizar um número adequado de repetições para estimar
corretamente o erro experimental.

Neste exemplo serão considerados:

- 15 corpos de prova de juta;
- 15 corpos de prova de abacaxi.

Total de ensaios: 30.

Esse número de repetições permite estimar a dispersão dos resultados e
realizar comparações estatísticas com maior confiabilidade.

A necessidade de um número maior de amostras está diretamente
relacionada à variabilidade intrínseca das fibras vegetais. Fatores como
idade da planta, condições de cultivo, extração das fibras e defeitos
naturais podem gerar diferenças significativas entre amostras
aparentemente semelhantes.

Por outro lado, caso o estudo envolvesse tecidos produzidos com fibras
sintéticas, como vidro, carbono ou polietileno, a variabilidade entre
amostras tenderia a ser menor devido ao processo industrial de
fabricação. Nesses casos, frequentemente é possível obter estimativas
confiáveis utilizando um número menor de repetições.

### Delineamento experimental

Será utilizado o Delineamento Inteiramente Casualizado (DIC).

Nesse delineamento, todos os corpos de prova possuem a mesma
probabilidade de serem ensaiados em qualquer posição da sequência
experimental.

A escolha do DIC é adequada porque:

- Existe apenas um fator em estudo;
- Os ensaios são realizados em laboratório;
- As condições experimentais são controladas.

### Casualização dos ensaios

A ordem dos ensaios não deve seguir a sequência dos tratamentos.

Em vez disso, recomenda-se realizar um sorteio da ordem experimental
para evitar que fatores externos influenciem sistematicamente os
resultados.

Exemplo:

AB → JU → JU → AB → AB → JU …

onde:

- JU = juta;
- AB = abacaxi.

```python
import random

tratamentos = ["JU"]*15 + ["AB"]*15

random.shuffle(tratamentos)

print("Ordem aleatória dos ensaios:\n")

for i, t in enumerate(tratamentos, start=1):
    print(f"Ensaio {i:02d}: {t}")
```

### Possíveis fontes de variabilidade

Mesmo utilizando o mesmo tratamento, os resultados podem variar devido a
fatores como:

- Variação natural das fibras;
- Defeitos locais no tecido;
- Diferenças na espessura dos fios;
- Alinhamento das fibras;
- Pequenas variações durante o ensaio.

Essas variações constituem o erro experimental e justificam a
necessidade de repetições.

### *Resumo do planejamento experimental*

**Problema:** Comparar o desempenho mecânico de tecidos de juta e
abacaxi.

**Fator:** Tipo de fibra vegetal.

**Níveis:** Juta e abacaxi.

**Tratamentos:** Tecido de juta e tecido de abacaxi.

**Unidade experimental:** Corpo de prova do tecido.

**Repetições:** 15 por tratamento.

**Delineamento:** Inteiramente casualizado (DIC).

**Variáveis resposta:** Resistência máxima, deformação, módulo de
elasticidade e tenacidade.
