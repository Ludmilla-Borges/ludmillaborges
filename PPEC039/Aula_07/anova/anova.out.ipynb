# **Análise de Variância (ANOVA)**

## **O que é a Analise de ANOVA?**

A Análise de Variância (ANOVA – Analysis of Variance) é uma técnica
estatística desenvolvida por Ronald Fisher com o objetivo de verificar
se existem diferenças estatisticamente significativas entre as médias de
três ou mais grupos. Trata-se de um dos métodos mais utilizados em
pesquisas científicas quando se deseja comparar populações ou
tratamentos distintos.

O princípio da ANOVA consiste em analisar a variabilidade dos dados,
dividindo-a em duas partes: a variabilidade existente dentro dos grupos
e a variabilidade existente entre os grupos. Quando a variabilidade
entre os grupos é significativamente maior que a variabilidade interna
de cada grupo, há evidências de que pelo menos um dos grupos apresenta
comportamento diferente dos demais.

## **A técnica baseia-se em duas hipóteses principais:**

1.  Hipótese Nula (H₀): As médias dos grupos são iguais.
2.  Hipótese Alternativa (H₁): Pelo menos uma das médias difere das
    demais.

## **A aplicação da ANOVA exige alguns pressupostos básicos:**

1.  Independência das observações.
2.  Normalidade dos dados.
3.  Homogeneidade das variâncias entre os grupos.

Caso esses pressupostos sejam atendidos, os resultados obtidos
apresentam maior confiabilidade estatística.

## **Analise de ANOVA - Olhando a Geração Cidadã de Dados**

### **Objetivo**

Avaliar se existem diferenças significativas no nível de compreensão
territorial entre grupos comunitários que apresentam diferentes níveis
de participação em processos de geração cidadã de dados geoespaciais.

### **Contextualização**

A geração cidadã de dados geoespaciais pode ser utilizada por diversas
comunidades para produzir informações sobre seus territórios, auxiliando
na identificação de problemas locais, no fortalecimento da participação
social e na articulação de ações comunitárias. Entretanto, uma questão
importante é verificar se diferentes níveis de participação nesses
processos influenciam a compreensão do território pelos participantes.

Para responder essa questão, pode-se utilizar a Análise de Variância
(ANOVA), técnica estatística que permite comparar as médias de três ou
mais grupos simultaneamente.

## **Exemplo - Olhando a Analise de Anova e a Geração Cidadã de Dados**

### **Dados Utilizados**

Foram considerados três grupos fictícios:

1.  Grupo A: Baixa participação em atividades de mapeamento.
2.  Grupo B: Participação moderada.
3.  Grupo C: Alta participação.

Cada participante recebeu uma nota de compreensão territorial variando
de 0 a 10.

```python
# Dados fictícios: Nota de compreensão territorial (0 a 10)

grupo_a = [4.2, 4.8, 5.1, 4.5, 5.0, 4.7]  # Baixa participação
grupo_b = [5.8, 6.1, 6.4, 6.0, 6.3, 5.9]  # Participação moderada
grupo_c = [7.2, 7.5, 8.0, 7.8, 8.1, 7.6]  # Alta participação

from scipy.stats import f_oneway

f, p = f_oneway(grupo_a, grupo_b, grupo_c)

print("Estatística F:", f)
print("P-valor:", p)
```

### **Explicando o Exemplo anterior.**

Nesse exemplo, foram comparados três grupos de participantes envolvidos
em atividades de geração cidadã de dados geoespaciais. **O Grupo A**
representa participantes com baixa participação em atividades de
mapeamento, **o Grupo B** participantes com participação moderada e **o
Grupo C** participantes com alta participação. O objetivo da análise é
verificar se existem diferenças significativas na compreensão
territorial entre os grupos. Como o p-valor é menor que 0,05, conclui-se
que existe diferença estatisticamente significativa entre as médias
analisadas.

### **Explicando o Exemplo a Seguir**

Neste exemplo, a ANOVA é utilizada para verificar se existem diferenças
significativas no tempo médio de deslocamento entre estudantes que
residem a diferentes distâncias da escola. Observa-se que os estudantes
do **Grupo A** apresentam tempos menores de deslocamento, enquanto os
estudantes do **Grupo C** possuem os maiores tempos. A análise busca
verificar se essas diferenças são estatisticamente significativas ou se
poderiam ocorrer apenas por variações aleatórias da amostra.

Caso o p-valor seja menor que 0,05, conclui-se que existe diferença
significativa entre os grupos, indicando que a distância da residência
influencia o tempo de deslocamento até a escola.

```python
# Dados fictícios

grupo_a = [8, 10, 12, 9, 11, 10]
grupo_b = [15, 18, 20, 17, 19, 16]
grupo_c = [25, 28, 30, 27, 29, 26]

from scipy.stats import f_oneway

f, p = f_oneway(grupo_a, grupo_b, grupo_c)

print("Estatística F:", f)
print("P-valor:", p)
```

## **Relação com o Projeto de Mestrado**

A aplicação da ANOVA permite verificar se a participação em processos de
geração cidadã de dados geoespaciais está associada a diferenças
significativas na compreensão do território. Essa análise dialoga
diretamente com a hipótese do projeto de mestrado, que pressupõe que a
apropriação de dados geoespaciais pelos atores locais fortalece
capacidades individuais e coletivas, contribuindo para uma melhor
leitura do território e para a organização de ações comunitárias.

## **Conclusão**

A ANOVA constitui uma ferramenta importante para comparar grupos e
identificar possíveis impactos decorrentes da participação em
iniciativas de geração cidadã de dados. No contexto desta atividade, a
técnica possibilita avaliar se diferentes níveis de envolvimento nos
processos de mapeamento estão associados a mudanças na compreensão
territorial dos participantes, fornecendo evidências quantitativas para
apoiar análises sobre participação social e fortalecimento comunitário.
